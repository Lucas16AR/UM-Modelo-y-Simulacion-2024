```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Definición de las ecuaciones de Lotka-Volterra
def lotka_volterra(y, t, a, p, b, q):
    liebres, zorros = y
    d_liebres_dt = a * liebres - b * liebres * zorros
    d_zorros_dt = -q * zorros + p * liebres * zorros
    return [d_liebres_dt, d_zorros_dt]

# Parámetros del modelo
a = 1.0 # Tasa de crecimiento de liebres
p = 0.1 # Tasa de crecimiento de zorros
b = 0.02 # Tasa de mortalidad de liebres
q = 0.3 # Tasa de mortalidad de zorros

# Condiciones iniciales
liebres_iniciales = 500
zorros_iniciales = 10
y0 = [liebres_iniciales, zorros_iniciales]

# Tiempo de simulación
t = np.linspace(0, 100, 1000)

# Resolución de las ecuaciones de Lotka-Volterra
sol = odeint(lotka_volterra, y0, t, args=(a, p, b, q))

# Extracción de las poblaciones de liebres y zorros
liebres = sol[:, 0]
zorros = sol[:, 1]
```

```python
# Gráfico de las poblaciones
plt.figure(figsize=(10, 6))
plt.plot(t, liebres, label='Liebres')
plt.plot(t, zorros, label='Zorros')
plt.xlabel('Tiempo')
plt.ylabel('Población')
plt.title('Simulación Depredador-Presa')
plt.legend()
plt.show()
```
