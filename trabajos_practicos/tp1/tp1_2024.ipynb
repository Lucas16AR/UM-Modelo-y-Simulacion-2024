### 1. Parametros iniciales

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función del modelo de Lotka-Volterra
def lotka_volterra(x, y, a, b, c, d):
    dx_dt = a*x - b*x*y
    dy_dt = -c*y + d*x*y
    return dx_dt, dy_dt

# Definir función para simular el modelo de Lotka-Volterra
def simulate_lotka_volterra(x0, y0, a, b, c, d, dt, steps):
    x = np.zeros(steps)
    y = np.zeros(steps)
    x[0] = x0
    y[0] = y0
    for i in range(1, steps):
        dx_dt, dy_dt = lotka_volterra(x[i-1], y[i-1], a, b, c, d)
        x[i] = x[i-1] + dx_dt * dt
        y[i] = y[i-1] + dy_dt * dt
        if x[i] < 0:
            x[i] = 0
        if y[i] < 0:
            y[i] = 0
    return x, y
```

### 2. Simulacion 1

```python
# Función para graficar las simulaciones y los diagramas de fase
def plot_simulation_and_phase(x, y, title):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    ax1.plot(x, label='Presas')
    ax1.plot(y, label='Depredadores')
    ax1.set_title(title)
    ax1.set_xlabel('Tiempo')
    ax1.set_ylabel('Población')
    ax1.legend()

    ax2.plot(x, y, label='Diagrama de fase')
    ax2.set_title('Diagrama de Fase')
    ax2.set_xlabel('Presas')
    ax2.set_ylabel('Depredadores')
    ax2.legend()

    plt.show()
```

```python
# Simulación con valores por defecto
x0 = 500 # Población inicial de presas
y0 = 10 # Población inicial de depredadores
a = 0.1 # Tasa de crecimiento de las presas
b = 0.02 # Tasa de mortalidad de las presas debido a los depredadores
c = 0.3 # Tasa de mortalidad de los depredadores
d = 0.01 # Tasa de crecimiento de los depredadores debido a las presas
dt = 0.01 # Paso de tiempo
steps = 1000 # Número de pasos de simulación

x, y = simulate_lotka_volterra(x0, y0, a, b, c, d, dt, steps)
plot_simulation_and_phase(x, y, 'Simulación con valores por defecto')
```

### 3. Simulacion 2

```python
x0_modificado = 300  # Población inicial de presas (modificado)
y0_modificado = 20   # Población inicial de depredadores (modificado)
a_modificado = 0.1   # Tasa de crecimiento de las presas (modificado)
b_modificado = 0.03  # Tasa de mortalidad de las presas debido a los depredadores (modificado)
c_modificado = 0.2   # Tasa de mortalidad de los depredadores (modificado)
d_modificado = 0.015 # Tasa de crecimiento de los depredadores debido a las presas (modificado)

x_modificado, y_modificado = simulate_lotka_volterra(x0_modificado, y0_modificado, a_modificado, b_modificado, c_modificado, d_modificado, dt, steps)
plot_simulation_and_phase(x_modificado, y_modificado, 'Segunda simulación con condiciones modificadas')
```
